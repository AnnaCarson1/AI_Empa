```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import re
import csv
```

```python
base_url = "https://www.unegui.mn/avto-mashin/-avtomashin-zarna/"
all_hrefs = []

```

```python
for i in range(1, 180):
  query_parameter = "?page="+str(i)
  url = base_url + query_parameter
  print(url)
  response = requests.get(url)
  if response.status_code != 200:
    print(response.status_code)
    print('request failed! ', url)
    continue
  soup = bs(response.text, "html.parser")
  li_list = soup.find_all("li", {"announcement-container"})
  for li in li_list:
    a = li.find("a")
    car_href = "https://www.unegui.mn" + a['href']
    print('-->', car_href)
    all_hrefs.append(car_href)
    print(len(all_hrefs))
```

```python
set_hrefs = set(all_hrefs)
print(len(all_hrefs), len(set_hrefs))
```

```python
class Car:
  def __init__(self, brand, model, price, engine, steering, mileage, fuel, transmission, drivetrain, entryYear, manufactureYear, exteriorColor, interiorColor, leasing, location, conditions, door, url, type1):
    self.brand = brand
    self.model = model
    self.price = price
    self.engine = engine
    self.steering = steering
    self.mileage = mileage
    self.fuel = fuel
    self.transmission = transmission
    self.drivetrain = drivetrain
    self.entryYear = entryYear
    self.manufactureYear = manufactureYear
    self.exteriorColor = exteriorColor
    self.interiorColor = interiorColor
    self.leasing = leasing
    self.location = location
    self.conditions = conditions
    self.door = door
    self.url = url
    self.type1 = type1
```

```python
def parseTitle(title):
    brand = ""
    model = ""
    match1 = None
    if title:
      match1 = re.search(r'\d{4}/\d{4}', title)
      if match1 is not None:
        title = title.replace(match1.group(), "")
      if title.startswith("Land Rover"):
          brand = "Land Rover"
          title = title.replace("Land Rover", "")
          words = re.split(r'\W+', title)
          result = [word for word in words if word not in '']
          for i in result:
              model += " "+ i
      else:
          words = re.split(r'\W+', title)
          result = [word for word in words if word not in '']
          brand = result[0]
          for i in result[1:]:
              model += " "+ i
          # print(result)

    return brand, model.strip()


```

```python
def numberParser(num):
  splitedStr = re.split(r' ', num)
  # print(splitedStr)
  return splitedStr[0].strip()
```

```python
# test url
# sets = []
# sets.append("https://www.unegui.mn/adv/7006980_toyota-prius-30-2011-2020/")
# sets.append("https://www.unegui.mn/adv/70151_dodge-ram-2005-2022/")
# sets.append("https://www.unegui.mn/adv/7020521_toyota-crown-2009-2017/")

id = 1
for url in set_hrefs:
# for url in sets:
  print(id, url)
  id += 1

  response = requests.get(url)
  if response.status_code != 200:
    print(f'request failed ${url}, status code ${response.status_code}')
    continue

  soup = bs(response.text, "html.parser")

  brand = ""
  model =""
  type1 = ""
  price = 0.0
  engine = 0.0
  steering = ""
  mileage = 0
  fuel = ""
  transmission = ""
  drivetrain = ""
  entryYear = 0
  manufactureYear = 0
  exteriorColor = ""
  interiorColor =""
  leasing = "YES"
  location  = ""
  url = ""
  conditions = "YES"
  door = 0

  title_tag = soup.find("h1", {"class": "title-announcement"})
  if title_tag is not None and title_tag.text is not None:
    title = title_tag.text.strip()
  else :
    title = "Автомашин Бусад, 0000/0000"

  brand, model = parseTitle(title)

  div = soup.find("div", {"class": "announcement-price__cost"})
  if div is not None:
    children = div.findChildren("meta", recursive=False)
  else:
    print("No matching element found")

  for i in children:
    if i.get("itemprop") == "price":
      price = float(i.get("content"))
      print(price)

  ul = soup.find("ul", {"class": "chars-column"})

  if ul is not None:
    list_el = ul.findChildren("li", recursive=False)
    for li in list_el:

      li_children = li.find_all()
      if len(li_children) == 2:
        if li_children[0].text.strip() == "Мотор багтаамж:":
          engine = float(numberParser(li_children[1].text.strip()))
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Хурдны хайрцаг:":
          transmission = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Хүрд:":
          steering = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Төрөл:":
          type1 = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Өнгө:":
          exteriorColor = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Үйлдвэрлэсэн он:":
          manufactureYear = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Орж ирсэн он:":
          entryYear =  li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Хөдөлгүүр:":
          fuel = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Дотор өнгө:":
          interiorColor = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Лизинг:":
          if "Лизинггүй" == li_children[1].text.strip():
            leasing = "NO"
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Хаяг байршил:":
          location = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Хөтлөгч:":
          drivetrain = li_children[1].text.strip()
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Явсан:":
          mileage = int(numberParser(li_children[1].text.strip()))
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Нөхцөл:":
          if li_children[1].text.strip() == "Дугаар аваагүй":
            conditions = "NO"
          elif li_children[1].text.strip() == " Дугаар авсан":
            conditions = "YES"
          # print(li_children[0].text.strip(), li_children[1].text.strip())
        elif li_children[0].text.strip() == "Хаалга:":
          door = int(li_children[1].text.strip())

    data.append(Car(brand, model, price, engine, steering, mileage, fuel, transmission, drivetrain, entryYear, manufactureYear,
                   exteriorColor, interiorColor, leasing, location, conditions, door, url, type1))

```
